```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data/HR_data.csv', index_col=0)
df.head()
```

```python
NON_CATEGORICAL_COLUMNS = ['HR_Mean', 'HR_Median', 'HR_std', 'HR_Min', 'HR_Max', 'HR_AUC']
df_ordinals = df[NON_CATEGORICAL_COLUMNS]
```

```python
df_corr = df_ordinals.corr()
df_corr
```

```python
## Scatterplot
sns.pairplot(df_ordinals)
```

```python
for col in df_ordinals.columns:
    sns.boxplot(y=col, data=df_ordinals)
    plt.show()
```
